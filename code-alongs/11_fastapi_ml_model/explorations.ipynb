```python
from pydantic import BaseModel, Field

class IrisInput(BaseModel):
    sepal_length: float = Field(gt=4, lt=8.5)
    sepal_width: float = Field(gt=1.8, lt=5)
    petal_length: float = Field(gt=0.8, lt=7.5)
    petal_width: float = Field(gt=0, lt=3)

IrisInput
```

```python
payload = IrisInput(sepal_length=4.2, sepal_width=2.5, petal_length=0.9, petal_width=1)
payload
```

```python
payload.model_dump()
```

```python
import pandas as pd 

pd.DataFrame(payload.model_dump(), index = [0])
```

```python
import httpx

url = "http://127.0.0.1:8000/api/iris/v1/predict"
payload = {"sepal_length": 5, "sepal_width": 4.5, "petal_length": 1.8, "petal_width": 1}
with httpx.Client(timeout=10) as client:
    response = client.post(url, json = payload)
    response.raise_for_status()

response
```
