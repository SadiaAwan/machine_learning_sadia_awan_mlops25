```python
import pandas as pd

df = pd.DataFrame(wine["data"], columns=wine["feature_names"])
df = pd.concat([df, pd.DataFrame(wine["target"], columns=["wine_class"])], axis=1)

df.head()
```

## Scikit learn steps

### 0. Divide into X and y

```python
X, y = df.drop("wine_class", axis=1), df["wine_class"]

X.head(2)
```

```python
y.head(2)
```

### 1. train|test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.5, random_state=42
)

X_train.shape, X_test.shape
```

### 2. scale dataset

- KNN very important to scale

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)

scaled_X_train = scaler.transform(X_train)
scaled_X_test = scaler.transform(X_test)

scaled_X_train.min(), scaled_X_train.max(), scaled_X_test.min(),scaled_X_test.max(),
```

### 3. KNN

```python
from sklearn.neighbors import KNeighborsClassifier

# for simplicity we'll go for default
model = KNeighborsClassifier(n_neighbors=5)
model.fit(scaled_X_train, y_train)
```

### 4. Predict on test data 

```python
y_pred = model.predict(scaled_X_test)
y_pred
```

```python
y_test.values
```

### 5. Evaluation

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm).plot()
print(classification_report(y_test, y_pred))
```
